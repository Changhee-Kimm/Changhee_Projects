
## 운영체재란 ? 


- Linux 계열 ubuntu 운영체제, 저 그림에서는 Kernel의 역할.

컴퓨터라는 대저택을 관리하는 집사!
집주인(최고 관리자)의 권한을 위임받아 보조 어플리에키션에게 연산자원을 분배하고 운영하고 관리하는 역할.

운영체제의 버전, 커널의 버전이 따로있다. 

EX) Window 2000 이후 : Windows 8, Version 6.3  --> 운영체제의 버전 8 커널 버전 6.3

### 셸

### 프로세스와 스레드
 - 프로그램 : 컴퓨터가 수행할 작업을 기록해 놓은 것
 - 프로세스 : 프로그램이 작업할 때 프로그램을 관리하는 단위로, 
 - 운영체제는 프로세스 단위로 메모리와 CPU 연산 등 컴퓨팅 자원을 분배함
 
 - 메모리 관리 : 운영체제는 가상 메모리 일부만 제공해 주는데, 다른 프로세스의 메모리를 훔쳐볼 수 없게 함. --> 프로세스 격리로 부르며 중요한 보안 기제. 프로세스간에 소통하기 위해서는 IPC 기법을 사용해야함.
 
 
 
 - CPU 자원 관리 : CPU 연산의 경우 코어 단위로 프로세스에 제공됨. 프로세스는 스레드 단위로 코어를 하나씩 사용할 수 있는데, 개발자는 더 많은 스레드를 사용하게 설계하여 다중 CPU 코어의 이점을 극대화 할 수 있다.
 
 프로세스 : 운영체제가 메모리를 할당하는 # 작업단위

 스레드 : 프로세스가 할당받은 메모리를 활용하는 # 실행단위

하나의 프로그램 안에서 여러개의 프로세스를 활용하는 것을 멀티프로세싱(Multiprocessing), 하나의 프로세스 안에서 여러개의 스레드를 활용하는 것을 멀티스레딩(Multithreading)이라고 합니다
