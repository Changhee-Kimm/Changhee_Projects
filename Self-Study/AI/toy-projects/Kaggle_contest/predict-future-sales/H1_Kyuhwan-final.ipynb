```python
import numpy as np
import pandas as pd
import os
from os.path import join

path = os.getenv('HOME')+'/Desktop/Changhee/changhee_git/Kaggle_contest/predict-future-sales/data/'

sales = pd.read_csv(path + 'sales_train.csv')
items = pd.read_csv(path + 'items.csv')
item_categories = pd.read_csv(path + 'item_categories.csv')
shops = pd.read_csv(path + 'shops.csv')
test = pd.read_csv(path + 'test.csv')
sample_sub = pd.read_csv(path + 'sample_submission.csv')
    
    
```

```python
sales.head()
```

```python
sales['date'] = pd.to_datetime(sales['date'], format ='%d.%m.%Y')
```

```python
dataset = sales.pivot_table(index = ['shop_id', 'item_id'], values = ['item_cnt_day'], columns = ['date_block_num'],
                           fill_value = 0, aggfunc = 'mean')
```

```python
dataset.reset_index(inplace = True)
```

```python
dataset.head()
```

```python
dataset = pd.merge(test, dataset, on = ['item_id', 'shop_id'], how = 'left')
```

```python
dataset.fillna(0, inplace = True)
```

```python
dataset.head()
```

```python
dataset.drop(['shop_id', 'item_id', 'ID'], inplace = True, axis =1)
dataset.head()
```

```python
# X 는 맨마지막 행을 제외한 모든 행을 유지
X_train = np.expand_dims(dataset.values[:,:-1], axis = 2)

# 맨 마지막 행은 라벨
y_train = dataset.values[:,-1:]

# 테스트를 위해 첫행을 제외한 모든 행을 유지

X_test = np.expand_dims(dataset.values[:,1:], axis =2)

print(X_train.shape, y_train.shape, X_test.shape)

```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
```

```python
model = Sequential()
model.add(LSTM(units = 64, input_shape = (33,1)))
model.add(Dropout(0.4))
model.add(Dense(3))

model.compile(loss = 'mse', optimizer ='adam', metrics = ['mean_squared_error'])
model.summary()
```

```python
model.fit(X_train, y_train, batch_size = 4096, epochs =50)
```

```python
prediction = model.predict(X_test)
```
